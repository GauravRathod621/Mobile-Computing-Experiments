### b) Write a menu-driven python code to implement a stack

```python
# Stack class
class Stack:
    """A class representing a Stack"""
    def __init__(self):
        self.st = [] # use a list
        
    def empty(self):
        return len(self.st) == 0
    
    def push(self, elem):
        # TODO
        # push elem on top of stack
        self.st.append(elem)
    
    def pop(self):
        # TODO
        # pop from top of stack
        # must return None if stack is empty
        # handle exceptions if any
        try:
            self.st.pop()
            
        except Exception as e:
            print(e)
    
    def peek(self):
        # TODO
        # return top of stack
        # must return None if stack is empty
        # handle exceptions if any
        try:
            return self.st[-1]
        except Exception as e:
            print(e)
    
    def search(self, elem):
        # TODO
        # return index of elem
        # must return None if stack is empty or element is not found
        # handle exceptions if any
        try:
            return self.st.index(elem)
        except Exception as e:
            print(e)
    def display(self):
        print("Stack", self.st)

# function
def menu():
    """Function to display a menu"""
    print("1. Push")
    print("2. Pop")
    print("3. Peek")
    print("4. Search element")
    print("0. Quit")

# create a stack
s = Stack()

# display menu
menu()
choice = int(input("Enter your choice: "))
while choice != 0:
    if choice == 1:
        elem = int(input("Enter an element: "))
        s.push(elem)
    elif choice == 2:
        print("Element removed:", s.pop())
    elif choice == 3:
        print("Element at top:", s.peek())
    elif choice == 4:
        elem = int(input("Enter an element: "))
        ind = s.search(elem)
        if ind == None:
            print("Element not found")
        else:
            print("Element found at index:", ind)
    elif choice == 0:
        break
    else:
        print("Invalid choice")

    s.display()

    menu()
    choice = int(input("Enter your choice: "))

```
